### NIFTY - Logarithmic Returns

```python
import numpy as np
import pandas as pd
import nsepy as nse
import matplotlib.pyplot as plt
```

```python
from nsepy import get_history
from datetime import date
```

```python
NIFTY = nse.get_history(symbol = 'NIFTY', index = True, start = date(2018, 1, 1), end = date(2023, 3, 23))['Close']
NIFTY
```

```python
(NIFTY/NIFTY.iloc[0]*100).plot()
```

### Log Returns

$$
ln(\frac{P_t}{P_{t-1}})
$$


```python
NIFTY['log_return'] = np.log(NIFTY/ NIFTY.shift(1))
print (NIFTY['log_return'])
```

```python
NIFTY['log_return'].plot(figsize=(8, 5))
plt.show()
```

### Calculate the average daily return

```python
log_return_d = NIFTY['log_return'].mean()
log_return_d
```

### Estimate the average annual return.

```python
log_return_a = NIFTY['log_return'].mean() * 250
log_return_a
```

Print the percentage version of the result as a float with 2 digits after the decimal point.

```python
print (str(round(log_return_a, 5) * 100) + ' %')
```
